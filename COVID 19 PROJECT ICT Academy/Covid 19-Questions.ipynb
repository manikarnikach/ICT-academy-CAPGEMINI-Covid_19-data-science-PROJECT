```python
# top 5 countries which have more number of covid cases (hint use df.nlargest)
top_counteries_cases=df.nlargest(5,'Total Cases')
top_counteries_cases
```

```python
row_number = 135
df.loc[row_number]
```

```python
#plot bar graph for top 5 countries
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.bar(top_counteries_cases['Country'], top_counteries_cases['Total Cases'])
plt.xlabel('Country')
plt.ylabel('Cases')
plt.title('Top 5 counteries by COVID Cases')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#Death percentage
#top 5 countries which have more number of covid  Death percentage 
most_countries_deathp=df.nlargest(5,'Death percentage')
most_countries_deathp
```

```python
#plot bar graph for top 5 countries which have more number of covid  Death percentage 
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.bar(most_countries_deathp['Country'], most_countries_deathp['Death percentage'])
plt.xlabel('Country')
plt.ylabel('Deaths')
plt.title('Least 5 counteries by COVID Cases')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#continent wise cases (hint use group by)
continent_cases =df.groupby('Continent')['Total Cases'].sum().reset_index()
continent_cases
```

```python
#draw bar plot for continent wise case
plt.figure(figsize=(10,6))
plt.bar(continent_cases['Continent'], continent_cases['Total Cases'])
plt.xlabel('Continent')
plt.ylabel('Cases')
plt.title('Continent-wise COVID Cases')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#find relationship between Population and total cases.Comment  on outliers
```

```python
#find relationship between Population and total cases.Comment  on outliers
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.scatter(df['Population'], df['Total Cases'])
plt.xlabel('Population')
plt.ylabel('Cases')
plt.title('Relationship between Population and total cases')
plt.xscale('log')
plt.yscale('log')
plt.tight_layout()
plt.show()
```

```python
# Spearman rank correlation heatmap.Comment on co-relation.
import seaborn as sns
correlation_matrix = df.corr(method='spearman') #calculates the Spearman rank correlation coefficients for all pairs of numerical columns in the DataFrame
# Create a heatmap using Seaborn
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, linewidths=.5)
plt.title('Spearman Rank Correlation Heatmap - COVID-19 Dataset')
plt.show()
```
